```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
df= pd.read_csv('Titanic.csv')
```

```python
df.head()
```

```python
df_new=pd.get_dummies(df,dtype='int')
```

```python
df_new
```

```python
frequency=apriori(df_new,min_support=0.5,use_colnames=True)
frequency
```
