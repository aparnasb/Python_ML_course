## Removing Outliers

```python
df
```

```python
df.boxplot('Salary')
plt.show()
```

```python
Q3=df['Salary'].quantile(0.75)
Q3
```

```python
Q1=df['Salary'].quantile(0.25)
Q1
```

```python
IQR= Q3-Q1
IQR
```

```python
Lower_bound= Q1-1.5*IQR
Lower_bound
```

```python
Upper_bound= Q3+1.5*IQR
Upper_bound
```

```python
df[(df['Salary']<Lower_bound) | (df['Salary']>Upper_bound)]
```

```python
df_no_outlier= df[(df['Salary']>Lower_bound) & (df['Salary']<Upper_bound)]
df_no_outlier
```

```python
df_no_outlier.boxplot('Salary')
plt.show()
```

## Concat and merge

```python
dict1={'names':['ajay','sanjay','raheem','sudha','loha'],'course':['python','python','java','R','R'],'marks':[7,8,2,1,5]}
dict2={'names':['basha','sam','suji'],'course':['java','DA','AI'],'marks':[4,7,8],'location':['mumbai','calcutta','orissa']}
```

```python
df1= pd.DataFrame(dict1)
df2= pd.DataFrame(dict2)
```

```python
df2
```

```python
df_new= pd.concat([df1,df2],ignore_index=True)
df_new
```

```python
dict3= {'names':['sanjay','raheem','sudha','raja'],'groups':[1,2,3,4]}
dict1={'names':['ajay','sanjay','raheem','sudha','loha'],'course':['python','python','java','R','R'],'marks':[7,8,2,1,5]}
df3= pd.DataFrame(dict3)
df1
```

```python
pd.merge(df1,df3,on='names',how='inner')
```

```python
pd.merge(df1,df3,on='names',how='left')
```

```python
pd.merge(df1,df3,on='names',how='right')
```

```python
df3
```

```python
pd.merge(df1,df3,on='names',how='inner')
```
