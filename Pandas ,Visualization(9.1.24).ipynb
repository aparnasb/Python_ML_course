## Crosstab

```python
pd.crosstab(index=df1['Type'],columns=df1['Weight'],margins=True)
```

## groupby

```python
df1.groupby(df1['Origin']).get_group('USA')
```

```python
df1.groupby(df1['Origin'])['Weight'].max()
```

```python
df1.groupby([df1['Origin'],df1['Type']])[['Weight','EngineSize']].agg([max,min,np.mean])
```

```python
df1[['Weight','EngineSize']].groupby([df1['Origin'],df1['Type']]).agg([max,min,np.mean])
```

```python
df1.pivot_table(index=['name_of_car','Type'],columns=['Origin','Model'],values='Weight',aggfunc=(['max','min','mean']))
```
