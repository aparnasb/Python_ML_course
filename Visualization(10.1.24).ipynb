```python
df['color'].value_counts().plot(kind='bar',color='red')
```

```python
df_color=pd.crosstab(index=df['color'],columns='count')
df_color
```

```python
a=df_color.plot(kind='bar',color='red',figsize=(7,7))
a.bar_label(a.containers[0],label_type='center');
```

```python
df_cut=pd.crosstab(index=df['color'],columns=df['cut'])
df_cut
```

```python
b=df_cut.plot(kind='bar',stacked=True)
b.bar_label(b.containers[0]);
```

```python
print(b.containers[0][0])
```

```python
b=df_cut.plot(kind='bar',stacked=True,figsize=(10,10))
for i in b.containers:
    for j in i:
        variable=[j.get_height() if j.get_height()>0 else '']
    b.bar_label(i,labels=variable)
```
